##  Урок 7. Многомерный статистический анализ. Линейная регрессия
### Задача№1
Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без.
****

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
zp=np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks=np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
plt.scatter(zp,ks)
plt.xlabel('Величина заработной платы')
plt.ylabel('Поведенческий кредитный скоринг', rotation=90)
plt.title('зависимость заработной платы и кредитного скоринга')
plt.show()
```

ПРедположим существование линейной зависимости с некоторым диапазонм погрешности, тогда функция будет
$ks=a+b\cdot zp$  , где коэффициент $b$ :  
$$b=\frac{\bar{xy}-\bar x\cdot\bar y}{\bar{x^2}-{\bar x}^2}$$

```python
b=(np.mean(zp * ks) - np.mean(zp) * np.mean(ks)) / (np.mean(zp**2) - np.mean(zp) ** 2)
print(f'Коэффициент \nb =  {b}')
```

Коэффициент $a$,расчитывается по формуле:  
$a=\bar y-b\cdot\bar x=\bar{ks}-b\cdot\bar{zp}$

```python
a=np.mean(ks)-b*np.mean(zp)
print(f'Коэффициент \na =  {a}')
```

```python
print(f'\nМожно построить прямую являющуюся корелляцией зависимости заработной платы и кредитного скоринга\n')
plt.scatter(zp,ks)
plt.plot(zp, 444.18+2.62*zp, c='r', label=r'$ks=444.18+2.62\cdot zp$')
plt.legend()
plt.xlabel('Величина заработной платы')
plt.ylabel('Поведенческий кредитный скоринг', rotation=90)
plt.title('Линейная зависимость')
plt.show()
```

Линейную взаимосвязь между заработной платой и кредитным скорингом можно описать функцией:  
$ks=444.18+2.62\cdot zp$  

```python
def mse(b, x, y):
    return np.sum((b*x-y)**2)/len(x)

alpha=1e-06
b=0.1
mse_min=mse(b,zp,ks)
i_min=1
b_min=b
for i in range(3000):
    b-=alpha*(2/len(zp))*np.sum((b*zp-ks)*zp)
    if i%100==0:
        print(f'Iteration = {i}, b={b}, mse={mse(b, zp,ks)}')
    if mse(b,zp,ks)>mse_min:
        print(f'Iteration = {i_min}, b={b_min}, mse={mse_min},\n \nДостигнут минимум, \nкоэффициент линейной регресии = {b}')
        break
    else:
        mse_min=mse(b,zp,ks)
        i_min=i
        b_min=b
```

## Задача№2
В каких случаях для вычисления доверительных интервалов и проверки статистических гипотез используется таблица значений функции Лапласа, а в каких - таблица критических точек распределения Стьюдента?
